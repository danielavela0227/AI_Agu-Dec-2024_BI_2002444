**Assignment 5. Decision Tree.** The student will train a decision tree model in Google Colab using the Python programming language. This model will use the "Breast Cancer Wisconsin (Diagnostic)" dataset; this dataset contains features computed from digitized images of breast cancer tumors and the corresponding diagnosis (malignant or benign).

Step 1. Load libraries.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

Step 2. Load data.

```python
breast_cancer = load_breast_cancer()
X = breast_cancer.data
Y = breast_cancer.target
```

```python
breast_cancer.target_names
```

Step 3. Data preprocessing.

```python
train_data, test_data, train_labels, test_labels = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Step 4. Model selection and training.

```python
clf = DecisionTreeClassifier()
clf.fit(train_data, train_labels)
```

```python
# Accuracy of the model in training
train_accuracy = clf.score(train_data, train_labels)
print(train_accuracy)
```

Step 5. Model testing.

```python
Y_pred = clf.predict(test_data)
```

```python
#Prediction 0
print(Y_pred[0])
print(test_labels[0])
```

```python
#Prediction 1
print(Y_pred[1])
print(test_labels[1])
```

```python
#Prediction 2
print(Y_pred[2])
print(test_labels[2])
```

```python
#Prediction 3
print(Y_pred[3])
print(test_labels[3])
```

```python
#Prediction 4
print(Y_pred[4])
print(test_labels[4])
```

```python
# General Accuarcy Report
accuracy = accuracy_score(test_labels, Y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(test_labels, Y_pred, target_names=breast_cancer.target_names))
```

Decision tree visualization.

```python
plt.figure(figsize=(20, 15))  # Increase figure size for better resolution
tree.plot_tree(clf, filled=True, feature_names=breast_cancer.feature_names, class_names=breast_cancer.target_names)
plt.title("Decision Tree for Breast Cancer Prediction")
plt.savefig("decision_tree.png", dpi=300)  # Increase DPI for better resolution
plt.show()
```
